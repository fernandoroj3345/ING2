La ingeniería financiera de un proyecto de software pronostica que será necesario invertir $1000 por mes durante 12 meses para llevarlo a cabo,asuma que el patrocinante pagara en forma mensual vencido por los flujos del proyecto. Si finaliza el proyecto en el mes 12 el patrocinante podrá lanzar negocios subyacentes por los que comenzará a recibir flujos de caja por un tiempo prolongado, para simplificar se puede considerar que el valor presente de esos flujos en ese momento equivale a $18000.- Para los cálculos financieros se utiliza una tasa efectiva mensual del 1%.¿Cual es el valor presente neto del proyecto supuesto su ejecución
ocurre según los planes?

```python
# Este script realiza el cálculo del VPN descontando las inversiones mensuales y restándolas del valor presente del flujo de caja que
# se espera al finalizar el proyecto en el mes 12.
# Datos del proyecto
inversion_mensual = 1000        # Inversión mensual en dólares
meses_inversion = 12            # Duración de la inversión en meses
flujo_vto = 18000               # Valor presente de los flujos futuros al mes 12
tasa_mensual = 0.01             # Tasa efectiva mensual (1%)

# Calcular el valor presente de las inversiones mensuales (pagos vencidos)
vpn_inversiones = sum([
    inversion_mensual / ((1 + tasa_mensual) ** i) 
    for i in range(1, meses_inversion + 1)
])

# Calcular el valor presente del flujo a partir del mes 12
vpn_flujo = flujo_vto / ((1 + tasa_mensual) ** meses_inversion)

# Calcular el Valor Presente Neto (VPN) del proyecto
vpn_proyecto = vpn_flujo - vpn_inversiones

print(f"El Valor Presente Neto (VPN) del proyecto es: ${vpn_proyecto:.2f}")
```
